```python
import requests
import math
import matplotlib.pyplot as plt
import shutil
from getpass import getpass
from PIL import Image, UnidentifiedImageError
from requests.exceptions import HTTPError
from io import BytesIO
from pathlib import Path
import torch
from torch.utils.data import DataLoader, Dataset
from torchvision.datasets import ImageFolder
from torchmetrics import Accuracy
from torchmetrics.classification import accuracy
import pytorch_lightning as pl
from huggingface_hub import HfApi, HfFolder, Repository, notebook_login
from transformers import ViTFeatureExtractor, ViTForImageClassification
import os
```

```python
train_data_dir = Path('file path/train')
validation_data_dir = Path('file path/val')
test_data_dir = Path('file path/test')
```

```python
ds = ImageFolder(train_data_dir)
val_ds = ImageFolder(validation_data_dir)
test_ds = ImageFolder(test_data_dir)

print("Length of train set: " + str(len(ds)))
print("Length of val set: " + str(len(val_ds)))
print("Length of test set: " + str(len(test_ds)))
```

```python
plt.figure(figsize=(20,10))
num_examples_per_class = 5
i = 1
for class_idx, class_name in enumerate(ds.classes):
  folder = ds.root / class_name
  for image_idx, image_path in enumerate(sorted(folder.glob('*'))):
    if image_path.suffix in ds.extensions:
      image = Image.open(image_path)
      plt.subplot(len(ds.classes), num_examples_per_class, i)
      ax = plt.gca()
      ax.set_title(class_name, size='xx-large', pad=5, loc='left', y=0, backgroundcolor='white')
      ax.axis('off')
      plt.imshow(image)
      i += 1
      if image_idx + 1 == num_examples_per_class:
        break
```

```python
# Preparing Labels
label2id = {}
id2label = {}

for i, class_name in enumerate(ds.classes):
    label2id[class_name] = str(i)
    id2label[str(i)] = class_name
```

```python
'''
Image Classification Collator
To apply our transforms to images, we'll use a custom collator class.
We'll initialize it using an instance of ViTFeatureExtractor and pass the collator instance to torch.utils.data.DataLoader's collate_fn kwarg.
'''
class ImageClassificationCollator:
    def __init__(self, feature_extractor):
        self.feature_extractor = feature_extractor

    def __call__(self, batch):
        encodings = self.feature_extractor([x[0] for x in batch], return_tensors='pt')
        encodings['labels'] = torch.tensor([x[1] for x in batch], dtype=torch.long)
        return encodings
```

```python
# Init Feature Extractor, Model, Data Loaders

vit_model = 'google/vit-large-patch16-224-in21k'
feature_extractor = ViTFeatureExtractor.from_pretrained(vit_model)
model = ViTForImageClassification.from_pretrained(
    vit_model,
    num_labels=len(label2id),
    label2id=label2id,
    id2label=id2label
)
collator = ImageClassificationCollator(feature_extractor)

train_loader = DataLoader(ds, batch_size=4, collate_fn=collator, num_workers=2, shuffle=True)
val_loader = DataLoader(val_ds, batch_size=4, collate_fn=collator, num_workers=2)
test_loader = DataLoader(test_ds, batch_size=1, collate_fn=collator, num_workers=2)

print("Number of train batch: " + str(len(train_loader)))
print("Number of val batch: " + str(len(val_loader)))
print("Number of test batch: " + str(len(test_loader)))
```

```python
class Classifier(pl.LightningModule):

    def __init__(self, model, lr: float = 2e-5, **kwargs):
        super().__init__()
        self.save_hyperparameters('lr', *list(kwargs))
        self.model = model
        self.forward = self.model.forward
        self.val_acc = Accuracy(
            task='multiclass' if model.config.num_labels > 2 else 'binary',
            num_classes=model.config.num_labels
        )

    def training_step(self, batch, batch_idx):
        outputs = self(**batch)
        self.log(f"train_loss", outputs.loss)
        return outputs.loss

    def validation_step(self, batch, batch_idx):
        outputs = self(**batch)
        self.log(f"val_loss", outputs.loss)
        acc = self.val_acc(outputs.logits.argmax(1), batch['labels'])
        self.log(f"val_acc", acc, prog_bar=True)
        return outputs.loss

    def test_step(self, batch, batch_idx):
        outputs = self(**batch)
        self.log(f"test_loss", outputs.loss)
        acc = self.val_acc(outputs.logits.argmax(1), batch['labels'])
        #print('Accuracy =', acc)
        #prediction = outputs.logits.max(1)
        #print('Prediction intensity =', prediction)
        self.log(f"test_acc", acc, prog_bar=True)
        return outputs.loss

    def predict_step(self, batch, batch_idx):
        outputs = self(**batch)
        pred = outputs.logits.argmax(1)
        max_intensity = outputs.logits.max(1)
        min_intensity = outputs.logits.min(1)
        #print("prediction:"+ str(pred)+ "baatch ID: " + str(batch_idx))
        #print("pred = ", pred)
        #print("type of pred", type(pred))
        #print(" pred intensity = ", pred_intensity)
        #print("type of intensity", type(pred_intensity))
        #print("\n")
        return pred, max_intensity, min_intensity


    def configure_optimizers(self):
        return torch.optim.Adam(self.parameters(), lr=self.hparams.lr)
```

```python
pl.seed_everything(42)
classifier = Classifier(model, lr=2e-5)
```

```python
from pytorch_lightning.callbacks import ModelCheckpoint

checkpoint_callback = ModelCheckpoint(dirpath='file path/algoBias',
                                      monitor = 'val_loss', save_top_k = 1, mode='min', verbose=True)
trainer = pl.Trainer(accelerator='gpu', devices=1, max_epochs=25, precision=16, callbacks=[checkpoint_callback])
```

```python
trainer.fit(classifier, train_loader, val_loader)
```

```python
trainer.test(classifier,test_loader)
```

**Test part**

```python
trainer = pl.Trainer(accelerator="cpu")    ##################  gpu
results = trainer.predict(classifier,test_loader, ckpt_path="file path/programs/vit_epoch=5-step=9000.ckpt")
```

```python
predicted = []
max_prediction_intensity = []
min_prediction_intensity = []
results_data = results[:]
for i in range(len(results_data)):
  temp1 = (results_data[i][0][0]).tolist()
  predicted.append(temp1)
  #print(temp1)
  temp2 = (results_data[i][1].values[0]).tolist()
  max_prediction_intensity.append(temp2)
  #print(temp2)
  temp3 = (results_data[i][2].values[0]).tolist()
  min_prediction_intensity.append(temp3)
  #print(temp3)
print("length of predicted list: " + str(len(predicted)))
```

```python
# label creation form test loader
import itertools

batches = iter(test_loader)
tmp_labels = []
for i in range(len(batches)):
  tmp_data = next(batches)
  key, data = tmp_data.items()
  #print(data[1])
  tmp_labels.append(data[1].tolist())

test_labels = list(itertools.chain(*tmp_labels))
print("length of groud truth list: "+ str(len(test_labels)))
```

```python
from mlxtend.evaluate import confusion_matrix
from mlxtend.plotting import plot_confusion_matrix
import numpy as np

class_names=['gan', 'real']
cm = confusion_matrix(y_target=test_labels,
                      y_predicted=predicted,
                      binary=False)
fig, ax = plot_confusion_matrix(conf_mat=cm,
                                show_normed=True,
                                cmap="YlGnBu",
                                colorbar=True)#,
                                #class_names=class_names)
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names,rotation=0)
plt.yticks(tick_marks, class_names)
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')
plt.show()
```

```python
from sklearn.metrics import classification_report

test_samples = 2000
print('\nTest Classification Report\n')
test_rpt = classification_report(test_labels, predicted, target_names=class_names)
print(test_rpt)
```
